```python
import numpy as np
import pandas as pd
import os 
```

```python
defaultdir = '/gpfs/radev/pi/saxena/aj764/Training_COOPERATION'
rootdir = f'{defaultdir}/behav/mag/'
```

```python
# all of the lever or mag files in Training COOPERATION / behav
lever_files = []
for (dirpath, dirnames, filenames) in os.walk(rootdir):
    lever_files.extend(filenames)
    break
```

```python
# all of the sessions folders in Training COOPERATION
date_fold = []
for (dirpath, dirnames, filenames) in os.walk(defaultdir):
    date_fold.extend(dirnames)
    break
```

```python
# date and animal id pairs for the lever/mag files
sorted_file = {}
for file in lever_files:
    parsed = file.split('_')
    date = parsed[0]
    animal = parsed[1]
    if date not in sorted_file.keys():
        sorted_file[date] = []
    sorted_file[date].append(animal)
```

```python
# prints the commands to copy the lever/mag files to the correct session folder
for key, value in sorted_file.items():
    for animal in value:
        if key in date_fold:
            vid_files = os.listdir(defaultdir + '/' + key)
            for vid in vid_files:
                # print(animal, vid)
                if animal in vid and 'mp4' in vid:
                    if not os.path.isdir(f'{defaultdir}/{key}/Behavioral'):
                        os.makedirs(f'{defaultdir}/{key}/Behavioral')
                    # print(key + animal, vid)
                    # print(f'cp behav/mag/{key + '_' + animal + '_mag.csv'} {key}/Behavioral/{key + '_' + animal + '_mag.csv'}')
                    
```
